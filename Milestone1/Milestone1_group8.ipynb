```python
# importing the relevant libraries
import numpy as np
import pandas as pd
import re  # to be able to clean the text using Regular Expression
import nltk
#nltk.download('punkt')
from nltk import word_tokenize 
from nltk.corpus import stopwords
#nltk.download('stopwords')
from nltk.stem import PorterStemmer

# reading the csv-file
data = pd.read_csv('news_sample.csv')

# inspecting the data
print(data.content)
print(data.shape)
data[:4]
```

```python
# TODO: More date formats needs to be recognized by the regex

def clean_text(text):
    text = str(text)
    text = text.lower() # lowercase
    text = re.sub("(\n+|\t+|\s{2,})", " ", text) # removing multible white spaces, tabs or new lines
    text = re.sub("[a-z]+ [0-9]{1,2}, [0-9]{4}", "<DATE>", text) # removes date in the format "month[letters] date, year"
    text = re.sub("(https?\:\/\/)?\w*.\w*.(com|net)[^\s\]]*", "<URL>", text) # urls replaced by <URL>
    text = re.sub("[0-9]+", "<NUM>", text) # numbers replaced by <NUM>, dates replaces by <DATE>
    text = re.sub("(\w+@\w+.com)", "<EMAIL>", text) # E-mail replaced by <EMAIL>
    return text

relString = 1
data["content"][relString]

# Right now we are only using clean_text on one string.
data_clean_homemade = clean_text(data["content"][relString])
print(data_clean_homemade)

print(clean_text(data["content"][relString])) # For testing purposes
```

```python
# to Process the text we will use the nltk library (https://www.nltk.org/)
# we have chosen to use our own data_clean_homemade, so we are able to remove dates etc. 

tokens = nltk.word_tokenize(data_clean_homemade) # to tokenize the text

print("Number of tokens in text: ", len(tokens))
#print(tokens[:10])

# stopwords
stopWords = set(stopwords.words('english'))

# removal of stopwords in tokens
tokens_without_stopWords = [w for w in tokens if not w in stopWords]
print("Number of tokens without stopwords: ", len(tokens_without_stopWords))

#print(tokens_without_Stopwords
reduction_rate = (len(tokens)-len(tokens_without_stopWords))/len(tokens)

print("We have removed %d stopwords from the text" %(len(tokens)-len(tokens_without_stopWords)))
print("Reduction rate: ", reduction_rate)

# Stemming
ps = PorterStemmer()

stem = []
for w in tokens_without_stopWords:
  stem.append(ps.stem(w))
print("Number of tokens in text without stopwords and after stemming: ", len(stem))

print("\nTo illustrate the stemming of the words")
print(tokens_without_stopWords[10:20])
print(stem[10:20])
```

In this task we were given a website, https://en.wikinews.org/wiki/Category:Politics_and_conflicts, and a list of letters, I through S.
The task was to collect data from all articles beginning with a letter in the list of letters and make this data fit for analysis.

We used that the articles on the website is sorted alphabetically and our list of letters does not loop around the alphabet (does not go from z to a at any place).

In order to gather data, we split the work into to different processes:


1.   Scrape overview-website for link to content-websites to scrape 
2.   Scrape data from content-websites to put into dataframe

Challenges we faced and solutions:


*   Accessing next page in the overviwe website. Solved by using selenium package
*   High runtime. Solved by stopping algorithm after last letter used
*   irregular use of ids. If statements to handle errors (Will be updated to try-with statements.)

```python
# Further improvements
# Start at site https://en.wikinews.org/w/index.php?title=Category:Politics_and_conflicts&from=I by clicking button

import requests
from bs4 import BeautifulSoup

RangeOfLetters = "ABCDEFGHIJKLMNOPRSTUVWZABCDEFGHIJKLMNOPRSTUVWZ"[8:8+10]
LetterList = list(RangeOfLetters)

# Create the webdriver to access the website
from selenium import webdriver
chrome_options = webdriver.ChromeOptions()
chrome_options.add_argument('--headless')
chrome_options.add_argument('--no-sandbox')
chrome_options.add_argument('--disable-dev-shm-usage')
wd = webdriver.Chrome('chromedriver',chrome_options=chrome_options)
driver =webdriver.Chrome('chromedriver',chrome_options=chrome_options)

# List of all article titles
articles = []

# Iterate through all pages on wikinews and find articles matching our assigned range of letters
url = "https://en.wikinews.org/wiki/Category:Politics_and_conflicts"
loop = True
index = 0
while loop:

  driver.get(url)
  response = requests.get(url)
  contents = response.text
  soup = BeautifulSoup(contents,'html.parser')
  res = []

  # Check if not at the last letter
  if index < (len(LetterList)-1):
    res = soup.find(id='mw-pages').find_all(title=re.compile("(^" + re.escape(LetterList[index]) + "\w*)"))
    res1 = soup.find(id='mw-pages').find_all(title=re.compile("(^" + re.escape(LetterList[index+1]) + "\w*)"))

    # If both the current and next letter is found on current page = end of current letter
    if len(res1) > 0:
      index = index + 1
    res = res + res1

  # Check if at last letter
  elif index == (len(LetterList)-1):
    res = soup.find(id='mw-pages').find_all(title=re.compile("(^" + re.escape(LetterList[index]) + "\w*)"))

  # Add all articles found current page to article list
  for x in res:
    underscore = x.text.replace(" ","_")
    articles.append("https://en.wikinews.org/wiki/" + str(underscore))

  # Goto next page if button can be found
  try:
    button = driver.find_element_by_link_text("next page")
    button.click()
    url = driver.current_url

  # If not = at last page, end loop
  except:
    loop = False
    
# Total number of articles matching the 10 assigned letters
print(len(articles))
```
